```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'results/test_12-10.csv'
parameter_ds = pd.read_csv(file_path)
```

```python
# Fix Dataset types
parameter_ds['position'] = parameter_ds['position'].astype('string')
parameter_ds['amt_num_features'] = parameter_ds['amt_num_features'].astype('string')
parameter_ds['stratify_by'] = parameter_ds['stratify_by'].astype('string')
```

```python
gk_dataset = parameter_ds[parameter_ds['position'] == 'GK']
```

## Window Size

```python
mse_stats = gk_dataset.groupby('window_size')['train_mse'].describe().drop(columns=['count'])
mae_stats = gk_dataset.groupby('window_size')['train_mae'].describe().drop(columns=['count'])
mse_stats[''] = ' ' 
pd.concat([mse_stats, mae_stats], axis=1, keys=["Train MSE", "Train MAE"])
```

```python
mse_stats = gk_dataset.groupby('window_size')['val_mse'].describe().drop(columns=['count'])
mae_stats = gk_dataset.groupby('window_size')['val_mae'].describe().drop(columns=['count'])

mse_stats[''] = ' ' 
pd.concat([mse_stats, mae_stats], axis=1, keys=["Validation MSE", "Validation MAE"])
```

```python
mse_stats = gk_dataset.groupby('window_size')['test_mse'].describe().drop(columns=['count'])
mae_stats = gk_dataset.groupby('window_size')['test_mae'].describe().drop(columns=['count'])
mse_stats[''] = ' ' 

pd.concat([mse_stats, mae_stats], axis=1, keys=["Test MSE", "Test MAE"])
```

A window of **size 6** has the best **test** and **validation** results.

Confidence of **60%**

```python
# Extract the required columns
x = parameter_ds['window_size']  # X-axis: window_size
y = parameter_ds['amt_num_features']  # Y-axis: amt_num_features
sizes = parameter_ds['test_mse']  # Sizes of points: test_mse

# Create the scatterplot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, s=sizes*5, c=sizes, cmap='cividis', alpha=0.7)  # Size scaled by test_mse

# Add a colorbar
plt.colorbar(scatter, label='Test MSE')

# Label the axes
plt.xlabel('Window Size')
plt.ylabel('Amount of Features')
plt.title('Test MSE by Window Size and Number of Features')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(parameter_ds['window_size'], parameter_ds['test_mse'], color='skyblue')
plt.xlabel('Window Size')
plt.ylabel('Test MSE')
plt.title('Test MSE vs Window Size')
plt.tight_layout()
plt.ylim(7, 11)

# Create the second bar plot (Number of Features vs Test MSE)
plt.figure(figsize=(10, 6))
plt.bar(parameter_ds['amt_num_features'], parameter_ds['test_mse'], color='lightcoral')
plt.xlabel('Amount of Features')
plt.ylabel('Test MSE')
plt.title('Test MSE vs Amount of Features')
plt.xticks(rotation=45)
plt.tight_layout()
plt.ylim(7, 11)

# Show both plots
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='window_size', y='test_mse', data=parameter_ds, palette='Set2')
plt.title('Test MSE Distribution by Window Size')
plt.xlabel('Window Size')
plt.ylabel('Test MSE')
plt.tight_layout()
plt.show()

# For 'amt_num_features' vs 'test_mse'
plt.figure(figsize=(10, 6))
sns.boxplot(x='amt_num_features', y='test_mse', data=parameter_ds, palette='Set2')
plt.title('Test MSE Distribution by Amount of Features')
plt.xlabel('Amount of Features')
plt.ylabel('Test MSE')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
