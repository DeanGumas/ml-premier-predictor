```python
import pandas as pd
file_path = 'results/test_12-10.csv'
parameter_ds = pd.read_csv(file_path)
```

```python
# Fix Dataset types
parameter_ds['position'] = parameter_ds['position'].astype('string')
parameter_ds['amt_num_features'] = parameter_ds['amt_num_features'].astype('string')
parameter_ds['stratify_by'] = parameter_ds['stratify_by'].astype('string')
```

```python
gk_dataset = parameter_ds[parameter_ds['position'] == 'GK']
```

## Window Size

```python
mse_stats = gk_dataset.groupby('window_size')['train_mse'].describe().drop(columns=['count'])
mae_stats = gk_dataset.groupby('window_size')['train_mae'].describe().drop(columns=['count'])
mse_stats[''] = ' ' 
pd.concat([mse_stats, mae_stats], axis=1, keys=["Train MSE", "Train MAE"])
```

```python
mse_stats = gk_dataset.groupby('window_size')['val_mse'].describe().drop(columns=['count'])
mae_stats = gk_dataset.groupby('window_size')['val_mae'].describe().drop(columns=['count'])

mse_stats[''] = ' ' 
pd.concat([mse_stats, mae_stats], axis=1, keys=["Validation MSE", "Validation MAE"])
```

```python
mse_stats = gk_dataset.groupby('window_size')['test_mse'].describe().drop(columns=['count'])
mae_stats = gk_dataset.groupby('window_size')['test_mae'].describe().drop(columns=['count'])
mse_stats[''] = ' ' 

pd.concat([mse_stats, mae_stats], axis=1, keys=["Test MSE", "Test MAE"])
```

A window of **size 6** has the best **test** and **validation** results.

Confidence of **60%**
